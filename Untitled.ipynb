```python
from elos.config import Config

# Load config, db handle
config = Config.load()
db = config.db()
```

```python
from elos.models import EventKind

events = [data for data in db.query(EventKind).select(
        {
            "owner_id": config.user_id,
            "name": "GrovePi Sensor Update"
        }
    ).limit("100000").execute()]

print("\n%d events" % len(events))

events
```
