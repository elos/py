##### Bootstrap elos, load our configuration and get a database handle

```python
from elos.config import Config

c = Config.load()
db = c.db()
```

##### We are going to watch as our user profile changes over time

```python
from elos.models import *
from websocket import create_connection

# We open a socket to recieve all model changes
changes = db.changes()

for change in changes:
    if change.record_kind == ProfileKind:
        profile = change.record_data.unmarshal(Profile)
        print(profile.marshal())
```
